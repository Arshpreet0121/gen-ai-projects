## Modes##
- r+ read and write
- a add new data 
- a+ 
- ab add new data in binary format
- ab+ create new file but doesnt overwrite
- w
- w+
- wb create file if not exist and overwrite if exist in binary
- wb+
- rb read in binary mode
- rb+ read and write in binary format
- x just to create file and dont want to do anything

```python
pass_count = 0
fail_count = 0

with open("newfile.txt", "r") as file:
    for line in file:
        line = line.strip().replace(",", " ")

        data = line.split()

        name = data[0]
        marks = int(data[1])

        if marks > 45:
            print(name, "-", marks, "- Pass")
            pass_count += 1
        else:
            print(name, "-", marks, "- Fail")
            fail_count += 1

print("\nTotal Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```
