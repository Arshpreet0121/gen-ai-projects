```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphones"],
    "Price": [60000, 800, 1500, 12000, 2500]
}

df = pd.DataFrame(data)
df.to_csv("products.csv", index=False)

df = pd.read_csv("products.csv")

average_price = np.mean(df["Price"])

print("Average Price:", average_price)

def above_average(df, avg):
    print("\nProducts priced above average:")
    for i in range(len(df)):
        if df["Price"][i] > avg:
            print(df["Product"][i], "-", df["Price"][i])

above_average(df, average_price)
```

```python
import pandas as pd
import numpy as np
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Temperature": [18, 25, 32, 28, 35]
}

df = pd.DataFrame(data)

df.to_csv("temperature.csv", index=False)
df = pd.read_csv("temperature.csv")

highest = np.max(df["Temperature"])
lowest = np.min(df["Temperature"])
average = np.mean(df["Temperature"])

print("Highest Temperature:", highest, "°C")
print("Lowest Temperature:", lowest, "°C")
print("Average Temperature:", average, "°C")

def classify(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"

print("\nTemperature Category:")
for i in range(len(df)):
    print(df["Day"][i], "-", df["Temperature"][i], "°C -", classify(df["Temperature"][i]))
```
