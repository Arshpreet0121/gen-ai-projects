```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = {
    "Study Hours": [2, 4, 6, 8, 10, 3, 5, 7, 9, 1],
    "Attendance": [70, 75, 80, 85, 90, 72, 78, 88, 95, 65],
    "Marks": [40, 50, 60, 70, 85, 45, 55, 72, 90, 35]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)

df = pd.read_csv("students.csv")

print("Missing Values:\n", df.isnull().sum())
df = df.dropna()

print("\nMean Marks:", np.mean(df["Marks"]))
print("Median Marks:", np.median(df["Marks"]))
print("Standard Deviation:", np.std(df["Marks"]))

plt.hist(df["Marks"], bins=5)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

plt.scatter(df["Study Hours"], df["Marks"])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

X = df[["Study Hours"]]
y = df["Marks"]

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)

print("\nPredicted Marks:")
print(pred)

plt.scatter(df["Study Hours"], df["Marks"], label="Actual Data")
plt.plot(df["Study Hours"], pred, color="red", label="Regression Line")
plt.title("Simple Linear Regression")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

columns = ["mpg", "cylinders", "displacement", "horsepower",
           "weight", "acceleration", "model_year",
           "origin", "car_name"]

df = pd.read_csv("auto-mpg.data",
                 delim_whitespace=True,
                 names=columns,
                 na_values="?")

df = df.dropna()

print("Descriptive Statistics:")
print(df.describe())

print("\nMean Mileage:", np.mean(df["mpg"]))
print("Median Mileage:", np.median(df["mpg"]))
print("Standard Deviation:", np.std(df["mpg"]))

plt.scatter(df["weight"], df["mpg"])
plt.title("Weight vs Mileage")
plt.xlabel("Car Weight")
plt.ylabel("Mileage (MPG)")
plt.show()

plt.hist(df["mpg"], bins=10)
plt.title("Histogram of Mileage")
plt.xlabel("Mileage (MPG)")
plt.ylabel("Frequency")
plt.show()

X = df[["displacement"]]      # Engine Size
y = df["mpg"]                 # Mileage

model = LinearRegression()
model.fit(X, y)

# Predicted values
pred = model.predict(X)

print("\nPredicted Mileage:")
print(pred)

# Regression Line
plt.scatter(df["displacement"], df["mpg"], label="Actual Data")
plt.plot(df["displacement"], pred, color="red", label="Regression Line")
plt.title("Engine Size vs Mileage")
plt.xlabel("Engine Size (Displacement)")
plt.ylabel("Mileage (MPG)")
plt.legend()
plt.show()
```
